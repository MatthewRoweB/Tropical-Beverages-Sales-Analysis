```python
## Personal Project 

import pandas as pd

df = pd.read_csv('Units_Types_Data.csv')
```

```python
df.head()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import pandas as pd

imputer = IterativeImputer(random_state=1, max_iter=1000, min_value=0)   # min_value = 0 emphasizes there is no negative values in imputation
imputer.fit(df)
df_imputed = imputer.transform(df)
new_df = pd.DataFrame(df_imputed, columns=df.columns)
```
